# **Task 1: Prediction Using Supervised Machine Learning**

In this task, we will predict the percentage of marks that a student is expected to score based upon the number of hours they studied. We use the available at http://bit.ly/w-data

**Importing Modules:**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Reading and preparing data**

```python
data = pd.read_csv('http://bit.ly/w-data')
print('Shape of the dataset is: ',data.shape)
data.head()
```

```python
#checking for any missing values

data.isnull().sum()
```

# Linear Regression :

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,1].values
```

```python
# spliting into test and train
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=50)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
m=reg.coef_
c=reg.intercept_
line=m*x+c
plt.scatter(x,y)
plt.plot(x,line)
plt.show()
```

```python
y_pred=reg.predict(x_test)
```

```python
actual_predicted=pd.DataFrame({'Target':y_test,'Predicted':y_pred})
actual_predicted
```

```python
sns.set_style('whitegrid')
sns.distplot(np.array(y_test-y_pred))
plt.show()
```

**What would be the predicted Score if a student studies for 9.25 hours/day?**

```python
h=9.25
s=reg.predict([[h]])
print("if a student studies for {} hours per day he/she will score {}% in exams.".format(h,s))
```

**Model Evaluation**

```python
from sklearn import metrics
from sklearn.metrics import r2_score
print("Mean Absolute Error:",metrics.mean_absolute_error(y_test,y_pred))
print("R2 Score:",r2_score(y_test,y_pred))
```
